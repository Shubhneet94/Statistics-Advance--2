#### Question 1: What is hypothesis testing in statistics? 

- Answer: Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on data from a sample. 

It helps us check whether a claim (hypothesis) about a population is likely to be true or not, using evidence from sample data. 

#### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis? 

- Answer: The null hypothesis is a formal statement that assumes there is no effect, no difference, or no relationship in the population. It serves as the default or starting point in hypothesis testing. Researchers test the null hypothesis to determine whether the observed data provides enough evidence to reject it. 

The alternative hypothesis is a statement that contradicts the null hypothesis. It represents the researcher’s claim or the possibility that there is an effect, a difference, or a relationship in the population. The goal of hypothesis testing is to gather evidence to support the alternative hypothesis by rejecting the null hypothesis. 

#### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test. 

Answer: The significance level, denoted by α, is a threshold value chosen by the researcher before conducting a hypothesis test. It represents the maximum probability of making a Type I error, which means rejecting the null hypothesis (H₀) when it is true. Role in Hypothesis Testing 

1. Comparison with p-value: 

   - After computing the test statistic, we calculate the p-value (the probability of observing the sample data if H₀ is true). 
   - If p-value ≤ α, we reject H₀ → evidence supports the alternative hypothesis (H₁). 
   - If p-value > α, we fail to reject H₀ → not enough evidence against the null hypothesis. 

2. Decision boundary: 

   - The significance level acts as the cut-off point for deciding whether results are statistically significant. 

3. Balance between risk and evidence: 

   - A smaller α (e.g., 0.01) makes the test more conservative (reduces false positives but increases chance of missing true effects). 
   - A larger α (e.g., 0.10) makes the test more lenient (easier to find significance but increases risk of false positives). 

#### Question 4: What are Type I and Type II errors? Give examples of each. 

- Answer: Type I Error (False Positive) 

    - Occurs when we reject the null hypothesis (H₀) even though it is true. 
    - In other words, we detect an effect or difference that does not exist. 
    - The probability of making a Type I error is equal to the significance level (α). 

Example: 

    - A medical test concludes that a healthy person has a disease. 
    - H₀: The person does not have the disease. 
    - H₁: The person has the disease. 
    - If the test incorrectly rejects H₀ (says the person is sick when they are healthy), that’s a Type I error. 

Type II Error (False Negative) 

    - Occurs when we fail to reject the null hypothesis (H₀) even though it is false. 
    - In other words, we miss an effect or difference that exists. 
    - The probability of making a Type II error is denoted by β. 

Example: 

    - A medical test concludes that a sick person is healthy. 
    - H₀: The person does not have the disease. 
    - H₁: The person has the disease. 
    - If the test fails to reject H₀ (says the person is healthy when they are sick), that’s a Type II error. 

#### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each. 

- Answer: Z-Test 

    - A Z-test is a statistical test used when: 
         - The population variance (σ²) or standard deviation (σ) is known. 
         - The sample size is large (n ≥ 30), so the Central Limit Theorem ensures the sample mean follows a normal distribution. 

    - It uses the standard normal distribution (Z-distribution). 

Example: 
 Testing whether the average weight of a large shipment of rice bags is 50 kg, when the population standard deviation is already known. 

T-Test 

    - A T-test is used when: 
       - The population variance (σ²) is unknown. 
       - The sample size is small (n < 30). 
    - It uses the student’s t-distribution, which is like the normal distribution but has heavier tails (to account for extra uncertainty when estimating σ from the sample). 
    - As the sample size increases, the t-distribution approaches the normal distribution, so large-sample t-tests ≈ z-tests. 

Example: 
 Testing whether the average exam score of a class of 15 students is 70, when the population standard deviation is unknown (so we rely on the sample standard deviation). 

 

#### Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of random samples

# Generate binomial distribution random numbers
binomial_data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.show()
```

#### Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] 

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Hypothesized population mean
mu_0 = 50

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # sample standard deviation
n = len(data)

# Z-statistic formula
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis (H0).")
else:
    print("Fail to Reject the Null Hypothesis (H0).")
```

#### Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)   # for reproducibility
mu, sigma = 50, 5    # true mean and std dev
n = 100              # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Compute 95% Confidence Interval for the mean
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # two-tailed t critical value
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval: [{:.3f}, {:.3f}]".format(ci_lower, ci_upper))

# Step 5: Plot histogram of the data
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"Upper 95% CI = {ci_upper:.2f}")

plt.title("Normal Distribution Sample with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

#### Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. 

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_zscores(data):
    """
    Function to calculate Z-scores for a dataset.
    Z = (x - mean) / std
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)   # sample std deviation
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
np.random.seed(42)
data = np.random.normal(50, 5, 100)  # normal distribution with mean=50, std=5, n=100

# Calculate Z-scores
z_scores = calculate_zscores(data)

# Print first few Z-scores
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
plt.legend()
plt.show()
```
